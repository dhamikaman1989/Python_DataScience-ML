## Clusteremap

```python
import seaborn as sns
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
flights.head(2)
```

```python
flight_pivot = flights.pivot_table(index = 'month',columns='year',values='passengers')
```

```python
sns.clustermap(flight_pivot,cmap = 'coolwarm')
```

## Grids

```python
# iris dataset

iris = sns.load_dataset('iris')
iris.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# PairGrid
iris_grid = sns.PairGrid(iris)    # creates grids for the numerical columns present in dataframe.
iris_grid.map(plt.scatter)
```

```python
# we can pass different plot types for the grid
sns.set_style('darkgrid')

iris_grid = sns.PairGrid(iris)

# create distribution plot for the diagonal...
iris_grid.map_diag(sns.distplot)

# create kdeplot for above the diagonal... 
iris_grid.map_upper(sns.kdeplot)

# create scattered plot for below the diagonal..
iris_grid.map_lower(plt.scatter)
```

## FacetGrid

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
# we define row & column for the grids first.. and, later we can map the other plots..

fg = sns.FacetGrid(data=tips,row='smoker', col='time')
fg.map(sns.distplot,'total_bill')  
    # if any function is needed more params - we can just add one after another...
```

## Regression Plots

```python
##lmplot - linear model plot

sns.lmplot(data= tips, x='total_bill',y='tip', hue='sex', markers=['o','v'])
```

```python
# or we can separate in two columns
sns.lmplot(data= tips, x='total_bill',y='tip', col ='sex')

# similarly can be row..
```
