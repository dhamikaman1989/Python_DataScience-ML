```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
# how to get categorical columns from df..

tip.select_dtypes(include = ['category']).columns
```

```python
tips.head(5)
```

```python
## Distribution plot.....
# Distribution of one variable..
```

```python
sns.distplot(tips['total_bill'])

# it creates kde (kernel destination estim plot) as well, which can be avoided by passing
# kde = False
```

```python
sns.set_style('darkgrid')

sns.distplot(tips['total_bill'], kde= False, bins = 30)  


# so its a histogram
```

```python
tips.columns
```

```python
## Jointplot
#  combines two distribution plots...
```

```python
sns.jointplot(data = tips, x = 'total_bill', y='tip')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y='tip', kind = 'hex')

# defaullt kind is scatter... we can change it to - reg (regression) - will show linear reg..
```

```python
sns.jointplot(data = tips, x = 'total_bill', y='tip', kind = 'reg')
```

```python
# when kind is kde
sns.jointplot(data = tips, x = 'total_bill', y='tip', kind = 'kde')
```

## Pairplot

```python
# it creates jointplots for each combinations of numerical columns present in the df...
# whereas between same columns - it plots histogram/kde
```

```python
sns.set_style('whitegrid')
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm') 
# for hue - we can pass categorical columns (not cont. data)
```

```python
## rugplot - only accepts one variable for plotting
# Similar to histogram - but in histogram, dashes are counted nd plotted
# marks dash for each data point
```

```python
sns.rugplot(tips['total_bill'])
```

```python
# kdeplot..

sns.kdeplot(tips['total_bill'])  
```

## Categorical plots

```python
tips.head()
```

```python
# barplot
# for x axis - pass the categorical columns; and for y axis - pass numerical column
# it will return the average value of y axis *default* can be changed
```

```python
sns.barplot(data=tips,x='sex',y='total_bill') 
```

```python
# can pass std for estimator field..

sns.barplot(data=tips,x='sex',y='total_bill', estimator= np.std)
```

```python
# what if we want to do counts?? 

# countplot  - only set the x value | y axis is already chose - i.e. count
```

```python
sns.countplot(data=tips,x='sex')
```

```python
# boxplot

# x axis will be categorical; y axis will be numerical..
```

```python
sns.boxplot(data = tips, x='day',y='total_bill')
```

```python
# violinplot
```

```python
fig = plt.figure(figsize = (10,5))
sns.violinplot(data = tips, x='day',y='total_bill')
```

```python
## stripplot
#  x axis - categorical; y axis - numerical
```

```python
sns.stripplot(data = tips, x ='day', y='total_bill')
```

```python
sns.stripplot(data = tips, x ='day', y='total_bill', hue='sex',split=True)
```

## catplot

```python
# catplot - default - stripplot

# can plot any categorical plot
#               e.g. : bar, box, violin,strip, count, point

sns.catplot(data=tips, x='day', y='total_bill', kind='bar',hue='sex')
```

## Matrix plots

```python
import seaborn as sns
%matplotlib inline
tips    = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head(5)
```

```python
flights.head()
```

```python
tip_corr = tips.corr()
```

```python
tip_corr
```

```python
## Heatmap...

# dataframe should've both index & columns with same names... so that these can be used for heatmap..
```

```python
sns.heatmap(tip_corr)
```

```python
# we can add annotation to get the exact value...  annot
# colormap : cmap = 'xxxx'

sns.heatmap(tip_corr,annot = True,cmap = 'coolwarm')
```

```python
# flights data

# create pivot table
```

```python
flight_pivot = flights.pivot_table(index='month',columns ='year', values= 'passengers')
```

```python
flight_pivot
```

```python
# plotting heatmap - with various line attributes.

sns.heatmap(flight_pivot, cmap = 'coolwarm', linewidths=1, linecolor = 'white')
```
