```python
x = np.linspace(0,5,11)
y = x **2
```

```python
# Functional way of plotting:

plt.plot(x,y,color='r')

plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Plot of y = x^2')

plt.show()
```

```python
plt.subplot(1,4,1)
plt.plot(x,y,'r')

plt.subplot(1,4,2)
plt.plot(y,x,'b')

plt.show()
```

```python
# Object oriented method
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.5,0.7])

axes.plot(x,y,'g')

axes.set_xlabel('X label')

axes.set_ylabel('Y label')

axes.set_title('Title')

plt.show()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig = plt.figure(figsize=(4,2))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
## Figsize and dpi
```

```python
fig, axes =  plt.subplots(figsize=(6,2))

axes.plot(x,y,'g')

#plt.margins(x=0)

axes.margins(x=0,y=0)

#axes.margins(y=0)
```

```python
fig, axes =  plt.subplots(nrows=2, ncols=1,figsize=(6,4))

axes[0].plot(x,y,'g')
axes[0].set_xlabel('X label')
axes[0].set_ylabel('Y label')
axes[0].margins(x=0,y=0)

axes[1].plot(y,x,'r')
axes[1].margins(x=0,y=0)

plt.tight_layout()
```

```python
# saving the plot...

fig.savefig('my_plot.png',dpi=200)
```
